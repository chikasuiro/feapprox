```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def contour_xi_x(xi):
    eta = np.linspace(-1, 1, 100)
    x = 5*xi*(xi-1)*eta*(eta-1) \
        -10*xi*(xi-1)*(eta-1)*(eta+1) \
        -25*(xi-1)*(xi+1)*eta*(eta-1) \
        +60*(xi-1)*(xi+1)*(eta-1)*(eta+1) \
        -35*(xi-1)*(xi+1)*eta*(eta+1) \
        +25*xi*(xi+1)*eta*(eta-1) \
        -55*xi*(xi+1)*(eta-1)*(eta+1) \
        +30*xi*(xi+1)*eta*(eta+1)
    return x
def contour_xi_y(xi):
    eta = np.linspace(-1, 1, 100)
    y = 5*xi*(xi-1)*eta*(eta-1) \
        -25*xi*(xi-1)*(eta-1)*(eta+1) \
        +35/2*xi*(xi-1)*eta*(eta+1) \
        -20*(xi-1)*(xi+1)*eta*(eta-1) \
        +55*(xi-1)*(xi+1)*(eta-1)*(eta+1) \
        -67/2*(xi-1)*(xi+1)*eta*(eta+1) \
        +15/2*xi*(xi+1)*eta*(eta-1) \
        -30*xi*(xi+1)*(eta-1)*(eta+1) \
        +20*xi*(xi+1)*eta*(eta+1)
    return y
```

```python
def contour_eta_x(eta):
    xi = np.linspace(-1, 1, 100)
    x = 5*xi*(xi-1)*eta*(eta-1) \
        -10*xi*(xi-1)*(eta-1)*(eta+1) \
        -25*(xi-1)*(xi+1)*eta*(eta-1) \
        +60*(xi-1)*(xi+1)*(eta-1)*(eta+1) \
        -35*(xi-1)*(xi+1)*eta*(eta+1) \
        +25*xi*(xi+1)*eta*(eta-1) \
        -55*xi*(xi+1)*(eta-1)*(eta+1) \
        +30*xi*(xi+1)*eta*(eta+1)
    return x
def contour_eta_y(eta):
    xi = np.linspace(-1, 1, 100)
    y = 5*xi*(xi-1)*eta*(eta-1) \
        -25*xi*(xi-1)*(eta-1)*(eta+1) \
        +35/2*xi*(xi-1)*eta*(eta+1) \
        -20*(xi-1)*(xi+1)*eta*(eta-1) \
        +55*(xi-1)*(xi+1)*(eta-1)*(eta+1) \
        -67/2*(xi-1)*(xi+1)*eta*(eta+1) \
        +15/2*xi*(xi+1)*eta*(eta-1) \
        -30*xi*(xi+1)*(eta-1)*(eta+1) \
        +20*xi*(xi+1)*eta*(eta+1)
    return y
```

```python
plt.plot(contour_xi_x(-1), contour_xi_y(-1), color='blue')
plt.plot(contour_xi_x(-1/2), contour_xi_y(-1/2), color='blue')
plt.plot(contour_xi_x(0), contour_xi_y(0), color='blue')
plt.plot(contour_xi_x(1/2), contour_xi_y(1/2), color='blue')
plt.plot(contour_xi_x(1), contour_xi_y(1), color='blue')

plt.plot(contour_eta_x(-1), contour_eta_y(-1), color='green')
plt.plot(contour_eta_x(-1/2), contour_eta_y(-1/2), color='green')
plt.plot(contour_eta_x(0), contour_eta_y(0), color='green')
plt.plot(contour_eta_x(1/2), contour_eta_y(1/2), color='green')
plt.plot(contour_eta_x(1), contour_eta_y(1), color='green')

plt.grid()
```
